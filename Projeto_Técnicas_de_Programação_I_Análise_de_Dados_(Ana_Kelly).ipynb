```python
bc = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Ada - Lets Code/Diversidade Tech/M3/Desafio/3.1-data-sheet-udemy-courses-business-courses.csv')
dc = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Ada - Lets Code/Diversidade Tech/M3/Desafio/3.1-data-sheet-udemy-courses-design-courses.csv')
mc = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Ada - Lets Code/Diversidade Tech/M3/Desafio/3.1-data-sheet-udemy-courses-music-courses.csv')
wd = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Ada - Lets Code/Diversidade Tech/M3/Desafio/3.1-data-sheet-udemy-courses-web-development.csv')

df = pd.concat([bc, dc, mc, wd])
df
```

```python
df.info()
```

```python
#cursos disponíveis na plataforma
df['course_id'].nunique()
```

2. Quais e quantos são os cursos que abordam o assunto: JavaScript?

```python
cursos_jsc = df[df['course_title'].str.contains('JavaScript', na=False)]
len(cursos_jsc)
```

3. Qual é o preço médio (coluna *price*) dos cursos oferecidos na plataforma?

```python
preco_medio = df['price'].mean()
print(f'{preco_medio:.2f}')
```

4. Quais são os cinco cursos com maior número de inscritos (coluna *num_subscribers*)?

```python
cursospop = df.sort_values('num_subscribers', ascending=False).head(5)
cursospop
```

5. Dos cinco cursos mais populares em termos de inscritos, mostre o **rate médio**, **máximo** e **mínimo** de cada um deles.

```python
meanPopRating = cursospop['Rating'].mean()
print(f'O rating médio entre os cursos mais populares é {meanPopRating:.2f}')

maxPopRating = cursospop['Rating'].max()
print(f'O rating máximo entre os cursos mais populares é {maxPopRating}')

minPopRating = cursospop['Rating'].min()
print(f'O rating mínimo entre os cursos mais populares é {minPopRating}')
```

6. Apresente os dez cursos mais visualizados na plataforma.
